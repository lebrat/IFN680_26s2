Normalization is a crucial preprocessing step that scales features to a similar range. This prevents features with large values from dominating the learning process.

- **Z-Score Normalization**: Rescales features to have a mean of 0 and a standard deviation of 1.
$$X_{normalized} = \frac{X - \mu}{\sigma}$$

$$X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}$$
